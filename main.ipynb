1. Import

```python
import pandas as pd 
import numpy as np
```

2. Loading Data


```python
data=pd.read_csv("data.csv",encoding="latin1")
x=data["no2"].dropna().values
print(x)
```

3. Transformation

```python
r=102316104
ar=0.05*(r%7)
br=0.3*((r%5)+1)
z=x+ar*np.sin(br*x)
```

4. Estimation of Parameter

```python
mean=np.mean(z)
sd=np.std(z)
lam=1/(2*sd**2)
c=np.sqrt(lam/np.pi)

print(" Mean = ",mean,"\n","Standard Deviation = ",sd,"\n","Lambda = ",lam,"\n","c = ",c)
```
